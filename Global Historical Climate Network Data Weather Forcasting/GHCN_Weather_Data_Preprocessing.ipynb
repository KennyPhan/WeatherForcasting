# 1. Importing Required Libraries

```python
import pandas as pd
```

# 2. Reading in the CSV data

```python
weather=pd.read_csv("MCI.csv",index_col="DATE")
weather
```

```python
# Check how many days are between the date range. Compare with rows.
from datetime import date
d0 = date(1973, 1, 1)
d1 = date(2023, 8, 23)
delta = d1 - d0
print('The number of days between the given range of dates is :')
print(delta.days+1)
```

# 3. Subsetting the data to core variables and dropping Null values

```python
# The documention details the core weather variables. (These variables have the lowest null values.)
core_weather = weather[["PRCP", "SNOW", "SNWD", "TMAX", "TMIN"]].copy()
core_weather.columns = ["precip", "snow", "snow_depth", "temp_max", "temp_min"]
core_weather
```

```python
# Check Null for core variables
core_weather.apply(pd.isnull).sum()
```

```python
# Dropping and checking null values
KC_Weather = core_weather.dropna()
KC_Weather.apply(pd.isnull).sum()
```

```python
KC_Weather.to_csv('KC_Weather.csv')
```
